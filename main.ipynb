```python
from model import NeuralNetwork
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the iris dataset
dataset = load_digits()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    dataset.data, dataset.target, test_size=0.25, random_state=42
)

# Normalize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
activation_functions = ["sigmoid", "tanh", "relu"]
```

```python
for activation_function in activation_functions:
    model = NeuralNetwork(
        input_size = X_train.shape[1],
        output_size = len(np.unique(y_train)),
        hidden_layers = [64],
        learning_rate = 0.01,
        activation_functions = [activation_function, "softmax"],
        loss_function = "categorical_crossentropy"
    )

    model.train(X_train, y_train, epochs=1000)

    print("Accuracy for {}: {}".format(activation_function, model.accuracy(X_test, y_test)))
```

```python
constant_initializer = ["ones", "zeros"]
```

```python
for initializer in constant_initializer:
    model = NeuralNetwork(
        input_size = X_train.shape[1],
        output_size = len(np.unique(y_train)),
        hidden_layers = [64],
        learning_rate = 0.01,
        activation_functions = ["relu", "softmax"],
        loss_function = "categorical_crossentropy",
        initializer = initializer
    )

    model.train(X_train, y_train, epochs=1000)

    print("Accuracy for {}: {}".format(initializer, model.accuracy(X_test, y_test)))
    
    
    loss, accuracy = model.history()

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(loss)
    plt.title("Loss evolution")
    plt.xlabel("epochs")
    plt.ylabel("loss")
    plt.subplot(1, 2, 2)
    plt.plot(accuracy)
    plt.title("Accuracy evolution")
    plt.xlabel("epochs")
    plt.ylabel("accuracy")
    plt.show()
```
